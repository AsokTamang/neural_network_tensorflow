```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input
from utils import load_coffee_data
```

**loading the datas**

```python
X,Y = load_coffee_data()
print('The first five training examples is:', X[:5])
print('The first five given target is:', Y[:5])
```

**Creating the mask**

```python
Y1 = Y.reshape(-1,)   #here we are reshaping for the plot, cause if we use 2D array in a plot , then the error of showing 3 indices occur
pos = (Y1 == 1)
neg = (Y1 == 0)
```

**plotting the data to check the relationship between input features and target variables**

```python
plt.scatter(X[pos,0],X[pos,1],marker='o',c='b',label='Positive pair')
plt.scatter(X[neg,0],X[neg,1],marker='x',c='r',label='Negative pair')
plt.title('Coffee Roasting')
plt.xlabel('Temperature(Celsius)')
plt.ylabel('Duration(minutes)')
plt.legend()
plt.show()
```

**Normalizing the training datas**

```python
nl_data = tf.keras.layers.Normalization(axis=-1)  #here we are using the normalization api of tensorflow, which normalizes the data feature-wise or column wise
nl_data.adapt(X)  #learning the mean and variance
X_normalized = nl_data(X)
print('The first five normalized datas is:', X_normalized[:5])
```

**Creating the neural network logistic model**

```python
tf.random.set_seed(1234)
model = Sequential(
    [
        Input(shape=(2,)),  #as there are two features in the training examples passed in the neural network
        Dense(3,activation='sigmoid',name='layer1'),
        Dense(1,activation='sigmoid',name='layer2')
    ]
)
model.summary()
```

**examining the weights and the bias of the models**

```python
layer1 = model.get_layer('layer1')
layer2=model.get_layer('layer2')
w1,b1 = layer1.get_weights()
w2,b2 = layer2.get_weights()
print(f'The initial parameters of the first layer is w:{w1} and b:{b1}')
print(f'The initial parameters of the second layer is w:{w2} and b:{b2}')
```

```python
Xt = np.tile(X_normalized,(1000,1))
Yt= np.tile(Y,(1000,1)) 
```

```python
#here we are using BinaryCrossentropy as loss function cause it is a logistic regression problem determining whether the coffee is good or not
model.compile(
     loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
)
model.fit(  #training the model
    Xt,Yt,
    epochs=10  #in each epochs the updation of the parameters or the gradient descent occur till all the samples are covered
)
```

```python
layer1 = model.get_layer('layer1')
layer2=model.get_layer('layer2')
final_w1,final_b1 = layer1.get_weights()
final_w2,final_b2 = layer2.get_weights()
print(f'The final parameters of the first layer is w:{final_w1} and b:{final_b1}')
print(f'The final parameters of the second layer is w:{final_w2} and b:{final_b2}')
```
