**forward propagation, this function dense takes the input or activation from the previous layer , does the computation of logistic expression then produces the activation of its own , so simply we are creating a hidden layer called dense**

```python
def dense(a_in,W,B):  #here a_in represents the activation passed from the previous layer, W and B represents the weight and bias of the current layer
    neurons = W.shape[1]  #number of neurons, as the shape of the weight is number of features by number of neurons
    a_out = np.zeros(neurons)
    for i in range(neurons):
        w = W[:,i]
        z = np.dot(a_in,w) + B[i]
        g = sigmoid(z)  #as this is the logistic regression , we must use the sigmoid function
        a_out[i] = g
    return a_out    
    
```

**here we are actually creating a neural network of 2 layers**

```python
def my_sequential(x,W1,B1,W2,B2):
    a1=dense(x,W1,B1)
    a2=dense(a1,W2,B2)
    return a2    
```
