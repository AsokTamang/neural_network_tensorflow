**creating a softmax function**

```python
def softmax_function(z):
    total_values = sum(np.exp(i) for i in z)
    n = len(z)
    a = np.zeros(n)
    for i in range(n):
        a[i] = np.exp(z[i]) / total_values
    return a    
```
