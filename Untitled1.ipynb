```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

**Neuron without activation in linearRegression Model**


```python
X_train = np.array([[1.0],[2.0]],dtype=np.float32)  #size in 1000 sqr.ft
Y_train = np.array([[300.0],[500.0]],dtype=np.float32)   #price of house in 1000 of dollars
X_train[0].reshape(1,1)
```

```python
plt.scatter(X_train,Y_train,marker='x',c='r',label='Data Points')
plt.legend()
plt.show()
```

**here we are creating a dense layer of unit 1 , which is connected to all the neurons in the prelayer**

```python
linear_layer = tf.keras.layers.Dense(units = 1 , activation = 'linear') 
linear_layer
```

**creating a layer in neural network having one neuron , with one of the training example**

```python
a1 = linear_layer(X_train[0].reshape(1,1)) 
a1
```

```python
w,b = linear_layer.get_weights()
print(f'The initial weight and the bias are : {w} and {b} respectively.')
```

**setting the parameters in the layer**

```python
set_w = np.array([[150]])
set_b = np.array([100])
linear_layer.set_weights([set_w,set_b])
linear_layer.get_weights()
```

```python
pred = np.dot(X_train[0],set_w) + set_b
print(pred)
a1 = linear_layer(X_train[0].reshape(1,1))   #again creating a layer in a neural network but this time with the new weight and bias
print(a1)
```

**passing the input in our neural network's layer**

**Neuron with sigmoid function**

```python
X_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.float32).reshape(-1,1)  # converting into 2-D Matrix
Y_train = np.array([0,  0, 0, 1, 1, 1], dtype=np.float32).reshape(-1,1)  # 2-D Matrix
```

**Creating the mask**

```python
pos = Y_train == 1
neg = Y_train == 0
```

```python
plt.scatter(X_train[pos],Y_train[pos],label='Positive',marker='o',c='b')
plt.scatter(X_train[neg],Y_train[neg],label='Negative',marker='x',c='r')
plt.legend()
plt.show()
```

```python
model = Sequential(
    [Input(shape=(1,)),  #assigning the number of features just before the first layer in the neural network, here its 1
        Dense(1,activation='sigmoid',name='L1')]  #here the number of neuron inside this layer is 1 and the the number of feature or number of input is 1 
)
```

```python
model.summary()
```

```python
logistic_model_layer = model.get_layer('L1')
print('layer of this logistic model is:', logistic_model_layer)
```

```python
w,b=logistic_model_layer.get_weights()
print(f'The initial parameters of this layer is: {w} and {b}')
```

```python
set_w = np.array([[2]])
set_b = np.array([-4.5])
logistic_model_layer.set_weights([set_w,set_b])
final_w,final_b = logistic_model_layer.get_weights()

```

```python
result_1 = model.predict()
```
